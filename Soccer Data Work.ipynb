```python
#Understat data - getting it per season and pivoting into a table that combines things.

import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None, 'display.max_rows', 10)

def data_build(url, league):
#url ='https://raw.githubusercontent.com/eddwebster/football_analytics/master/data/understat/raw/metadata/premier-league/metadata_premier_league_2021.csv'
    if url == 'https://raw.githubusercontent.com/eddwebster/football_analytics/master/data/understat/raw/metadata/ligue-1/metadata_ligue1_1415_.csv':
        year = url.split('_')[-2]
        year1 = year[:2]
        year2 = year[2:4]
        year_list = []
        year_list.append(year1)
        year_list.append(year2)
        year = '-'.join(year_list)
    elif url != 'https://raw.githubusercontent.com/eddwebster/football_analytics/master/data/understat/raw/metadata/ligue-1/metadata_ligue1_1415_.csv':
        year = url.split('_')[-1].split('.')[0]
        year1 = year[:2]
        year2 = year[2:4]
        year_list = []
        year_list.append(year1)
        year_list.append(year2)
        year = '-'.join(year_list)
    df = pd.read_csv(url, index_col = 0)
    df = df.rename(columns={'games':f'{year}_games', 'time':f'{year}_mins_played', 'goals':f'{year}_goals', 
                            'xG':f'{year}_xG',
                           'assists':f'{year}_assists', 'xA':f'{year}_xA', 'shots':f'{year}_shots',
                           'key_passes':f'{year}_key_passes', 'yellow_cards':f'{year}_yellow_cards',
                           'red_cards':f'{year}_red_cards', 'position':f'{year}_position', 'team_title':f'{year}_team',
                           'npg':f'{year}_npg', 'npxG':f'{year}_npxG', 'xGChain':f'{year}_xGChain', 
                            'xGBuildup':f'{year}_xGBuildup'})
    df.insert(4, f'{year}_goals_per_90', (df[f'{year}_goals']/df[f'{year}_mins_played'])*90)
    df.insert(6, f'{year}_xG_per_90', (df[f'{year}_xG']/df[f'{year}_mins_played'])*90)
    df.insert(8, f'{year}_assists_per_90', (df[f'{year}_assists']/df[f'{year}_mins_played'])*90)
    df.insert(10, f'{year}_xA_per_90', (df[f'{year}_xA']/df[f'{year}_mins_played'])*90)
    df.insert(12, f'{year}_shots_per_90', (df[f'{year}_shots']/df[f'{year}_mins_played'])*90)
    df.insert(14, f'{year}_key_passes_per_90', (df[f'{year}_key_passes']/df[f'{year}_mins_played'])*90)
    df.insert(5, f'{year}_npg_per_90', (df[f'{year}_npg']/df[f'{year}_mins_played'])*90)
    df.insert(8, f'{year}_npxG_per_90', (df[f'{year}_npxG']/df[f'{year}_mins_played'])*90)
    df.insert(13, f'{year}_xGChain_per_90', (df[f'{year}_xGChain']/df[f'{year}_mins_played'])*90)
    df.insert(14, f'{year}_xGBuildup_per_90', (df[f'{year}_xGBuildup']/df[f'{year}_mins_played'])*90)
    df = df[['player_name', f'{year}_games', f'{year}_mins_played', f'{year}_goals', f'{year}_goals_per_90', 
             f'{year}_npg', f'{year}_npg_per_90',
            f'{year}_xG', f'{year}_xG_per_90', f'{year}_npxG',f'{year}_npxG_per_90', f'{year}_assists', 
             f'{year}_assists_per_90',
            f'{year}_xA', f'{year}_xA_per_90', f'{year}_xGChain', f'{year}_xGChain_per_90', f'{year}_xGBuildup', 
             f'{year}_xGBuildup_per_90',
            f'{year}_shots', f'{year}_shots_per_90', f'{year}_key_passes', f'{year}_key_passes_per_90',
            f'{year}_yellow_cards', f'{year}_red_cards', f'{year}_team']]
    df[f'{year}_league'] = league
    df.to_csv(f"/Users/matthewbeale/Documents/Python/Soccer/League CSV's/{year}_{league}.csv", index=True)
    print(f'{league} in {year} csv created')
    return df

#data_build(url='https://raw.githubusercontent.com/eddwebster/football_analytics/master/data/understat/raw/metadata/premier-league/metadata_premier_league_2021.csv',
         # league = 'BPL')
```

```python
df_14_15 = pd.concat([pd.read_csv(file_name_list[0]), pd.read_csv(file_name_list[1]), 
                     pd.read_csv(file_name_list[2]),
                     pd.read_csv(file_name_list[3]), pd.read_csv(file_name_list[4])])
def fix_duplicates_df(df, year):
    p = '_per_90'
    df1 = df[df.duplicated(['player_name', 'id'], keep=False)].sort_values(by=['player_name', 
                                                                                 f'{year}_mins_played'], 
                                                                            ascending = [True, False])
    df1 = df1.drop(columns = [f'{year}_goals{p}', f'{year}_npg{p}',
                           f'{year}_xG{p}', f'{year}_npxG{p}', f'{year}_assists{p}',
                           f'{year}_xA{p}', f'{year}_xGChain{p}', f'{year}_xGBuildup{p}',
                           f'{year}_shots{p}', f'{year}_key_passes{p}'])
    team_only_df = df1[['id', 'player_name', f'{year}_team', f'{year}_league']]
    team_only_df = team_only_df.drop_duplicates(subset = ['id', 'player_name'])
    stats_only_df = df1.drop(columns = [f'{year}_team', f'{year}_league'])
    total_year_df = stats_only_df.groupby(['id', 'player_name']).sum().reset_index()
    total_year_df.insert(4, f'{year}_goals_per_90', 
                         (total_year_df[f'{year}_goals']/total_year_df[f'{year}_mins_played'])*90)
    total_year_df.insert(6, f'{year}_xG_per_90', 
                         (total_year_df[f'{year}_xG']/total_year_df[f'{year}_mins_played'])*90)
    total_year_df.insert(8, f'{year}_assists_per_90', 
                         (total_year_df[f'{year}_assists']/total_year_df[f'{year}_mins_played'])*90)
    total_year_df.insert(10, f'{year}_xA_per_90', 
                         (total_year_df[f'{year}_xA']/total_year_df[f'{year}_mins_played'])*90)
    total_year_df.insert(12, f'{year}_shots_per_90', 
                         (total_year_df[f'{year}_shots']/total_year_df[f'{year}_mins_played'])*90)
    total_year_df.insert(14, f'{year}_key_passes_per_90', 
                         (total_year_df[f'{year}_key_passes']/total_year_df[f'{year}_mins_played'])*90)
    total_year_df.insert(5, f'{year}_npg_per_90', 
              (total_year_df[f'{year}_npg']/total_year_df[f'{year}_mins_played'])*90)
    total_year_df.insert(8, f'{year}_npxG_per_90', 
              (total_year_df[f'{year}_npxG']/total_year_df[f'{year}_mins_played'])*90)
    total_year_df.insert(13, f'{year}_xGChain_per_90', 
              (total_year_df[f'{year}_xGChain']/total_year_df[f'{year}_mins_played'])*90)
    total_year_df.insert(14, f'{year}_xGBuildup_per_90', 
              (total_year_df[f'{year}_xGBuildup']/total_year_df[f'{year}_mins_played'])*90)
    total_year_df = total_year_df[['id', 'player_name', f'{year}_games', f'{year}_mins_played', f'{year}_goals', f'{year}_goals_per_90', 
             f'{year}_npg', f'{year}_npg_per_90',
            f'{year}_xG', f'{year}_xG_per_90', f'{year}_npxG',f'{year}_npxG_per_90', f'{year}_assists', 
             f'{year}_assists_per_90',
            f'{year}_xA', f'{year}_xA_per_90', f'{year}_xGChain', f'{year}_xGChain_per_90', f'{year}_xGBuildup', 
             f'{year}_xGBuildup_per_90',
            f'{year}_shots', f'{year}_shots_per_90', f'{year}_key_passes', f'{year}_key_passes_per_90',
            f'{year}_yellow_cards', f'{year}_red_cards']].sort_values(['player_name'])
    duplicate_df = pd.merge(total_year_df, team_only_df)
    duplicate_df[f'{year}_transferred'] = 'Yes'
    df2 = df.drop_duplicates(subset = ['id', 'player_name'], keep = False)
    overall_df = pd.concat([df2, duplicate_df])
    overall_df = overall_df.loc[overall_df[f'{year}_mins_played']>450].sort_values(by=['player_name'])
    overall_df = overall_df.rename(columns = {f'{year}_team':'team', f'{year}_transferred':'transferred'})
    overall_df.loc[overall_df.team.str.contains(','), 'transferred'] = 'Yes'
    overall_df['transferred'].fillna('No', inplace = True)
    overall_df = overall_df.rename(columns = {'team': f'{year}_team', 'transferred':f'{year}_transferred'})
    overall_df.to_csv(f"/Users/matthewbeale/Documents/Python/Soccer/Overall CSV's/{year}.csv", index = False)
    return overall_df



fix_duplicates_df(df_14_15, '14-15')

```

```python
#Twitter data fetching
import twint
import nest_asyncio
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
import statistics
final_sentiment_scores_list = []
#from path import exists
nest_asyncio.apply()
roberta = 'cardiffnlp/twitter-roberta-base-sentiment'
model =  AutoModelForSequenceClassification.from_pretrained(roberta)
tokenizer = AutoTokenizer.from_pretrained(roberta)
labels = ['Negative', 'Neutral', 'Positive']
final_scores = []
print('Packages imported.')

#Getting twitter data
#Will need year as an input to find teams to deal with potential problems with 
#duplicate player names
df = pd.read_csv("/Users/matthewbeale/Documents/Python/Soccer/final_csv.csv")
df =df[['id', 'player_name']]
df = df.sample(10)

# Custom player list for english stuff
player_list = df['player_name'].tolist()
player_list = ['Harry Kane', 'Mason Mount', 'Jack Grealish', 'Nick Pope', 'Eric Dier', 'Harry Maguire']
player_id_list = df['id'].tolist()
player_id_list = [647, 7768, 675, 5552, 643, 1687]
player_dict = dict(zip(player_id_list, player_list))
for key, value in player_dict.items():
    the_id = key
    player=value
    c = twint.Config()
    c.Search = player
    c.Store_csv = True
    #Will be problems with PLayer Names being duplicate CV's as well
    c.Output = f'/Users/matthewbeale/Documents/Python/Soccer/Twitter CSVs/19_20_{player}_{the_id}.csv'
    c.Since = '2019-08-01'
    c.Until = '2020-07-31'
    c.Language = 'en'
    c.Limit=10
    twint.run.Search(c)
    df = pd.read_csv(f'/Users/matthewbeale/Documents/Python/Soccer/Twitter CSVs/19_20_{player}_{the_id}.csv')
    eng_df = df_where(df, 'language', 'en')
    tweet = eng_df['tweet'].tolist()
    for t in tweet:
        tweet_words = []
        for word in t.split(' '):
            if word.startswith('@') and len(word)>1:
                word = '@user'
            if word.startswith('https'):
                word = 'http'
            tweet_words.append(word)
    
        tweet_proc = ' '.join(tweet_words)

#Perform analysis
        encoded_tweet = tokenizer(tweet_proc, return_tensors = 'pt')
        output = model(**encoded_tweet)

        scores = output[0][0].detach().numpy()
        scores = softmax(scores)
        if scores[1]>scores[0] and scores[1] > scores[2]:
            final_score = 0
            #print(final_score)
        else:
            final_score = scores[2] - scores[0]
            #print(final_score)
            final_scores.append(final_score)
    #print(final_scores)
    sentiment_score = statistics.mean(final_scores)
    final_sentiment_scores_list.append(sentiment_score)
    #print(f"{player}'s Sentiment Score = {sentiment_score}")

sentiment_df = pd.DataFrame({'Player_Name':player_list, 'Player_ID':player_id_list, 
                             '2019_2020_Sentiment_Score':final_sentiment_scores_list})
sentiment_df
```
